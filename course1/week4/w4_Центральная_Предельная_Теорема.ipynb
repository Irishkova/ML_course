```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import math
```

# распределение хи-квадрат с k степенями свободы. K = 6

```python
ourChi2 = sts.chi2(6)
```

Сгенерируйте из него выборку объёма 1000

```python
# выборка на 1000 элементов
sample = ourChi2.rvs(1000)
```

постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины

```python
# на отрезке [0, 30] рисуется...
x = np.linspace(0, 30, 100)
# ...гистограмма
pdf = ourChi2.pdf(x)
plt.hist(sample, density=True, label='гистограма выборки') #  (чтобы величины были в одном масштабе, не забудьте выставить у гистограммы значение параметра normed=True)

plt.plot(x, pdf, label='теоретическая плотность')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Для этого при трёх и более значениях n (например, 5, 10, 50) сгенерируйте 1000 выборок объёма n и постройте гистограммы распределений их выборочных средних.

# Оценим распределение выборочного среднего случайной величины при разных объёмах выборок 


```python
def sample(n):
    sampleMeans = []
    for rvn in range(1000):
        sampleMeans.append(ourChi2.rvs(n).mean())
        
    # рисуем:
    plt.hist(sampleMeans, 20, density=True, range=[0, 12], label='Гистограма выборочного среднего')
    # вычисляется pdf нормального распределения с данными параметрами (каждый раз одинково)
    norm_rv = sts.norm(6, np.sqrt(2.*6./n))
    x = np.linspace(0, 12, 100)
    pdf = norm_rv.pdf(x)
    # и вырисовывается на графике
    plt.plot(x, pdf, label='Теоретическое среднее')
    # легенда графика
    plt.legend()
    plt.title("Объем выборки %d" % n)
    plt.ylabel('$f(x)$')
    plt.xlabel('Выборочное среднее $x$')
    plt.show()
```

# при n = 5

```python
sample(5)
```

# при N = 10

```python
sample(10)
```

# при N = 50

```python
sample(50)
```

При увеличении размера выборки n, распределение средних значений начинает сходиться к 6, среднеквадратичное отклонение уменьшается => увеличении n точность апроксимации растет.

При этом расчетное нормальное распределение все более совпадает с гистограммой.
