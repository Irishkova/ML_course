```python
import sklearn
import numpy as np
import pandas as pd
from sklearn import *
```

```python
digits = sklearn.datasets.load_digits()
```

```python
X = digits['data']
y = digits['target']
```

```python
#1
classifier1 = sklearn.tree.DecisionTreeClassifier(random_state=0)
classifier1_score = sklearn.model_selection.cross_val_score(classifier1, X, y, cv=10, n_jobs=4).mean()
classifier1_score
```

```python
#2
classifier2 = sklearn.ensemble.BaggingClassifier(base_estimator=classifier1, n_estimators=100, random_state=0)
classifier2_score = sklearn.model_selection.cross_val_score(classifier2, X, y, cv=10, n_jobs=4).mean()
classifier2_score
```

```python
#3
classifier3 = sklearn.ensemble.BaggingClassifier(base_estimator=classifier1, n_estimators=100, random_state=0, max_features=0.33)
classifier3_score = sklearn.model_selection.cross_val_score(classifier3, X, y, cv=10, n_jobs=4).mean()
classifier3_score
```

```python
#4
classifier4 = sklearn.tree.DecisionTreeClassifier(random_state=0, splitter = 'random', max_features=0.33)
classifier5 = sklearn.ensemble.BaggingClassifier(base_estimator=classifier1, n_estimators=100, random_state=0)
classifier5_score = sklearn.model_selection.cross_val_score(classifier5, X, y, cv=10, n_jobs=4).mean()
classifier5_score
```

```python
#5
classifier6 = sklearn.ensemble.RandomForestClassifier()
classifier6_score = sklearn.model_selection.cross_val_score(classifier6, X, y, cv=10, n_jobs=4).mean()
classifier6_score
```

```python
#n_estimators
classifier_score_list = []
n_list = range(10,500,10)
for n in n_list:
    classifier = sklearn.ensemble.RandomForestClassifier(n_estimators=n)
    classifier_score_list.append(sklearn.model_selection.cross_val_score(classifier, X, y, cv=10, n_jobs=4).mean())
plot(n_list, classifier_score_list)
```

```python
#max_depth
classifier_score_list = []
d_list = range(1,20)
for d in d_list:
    classifier = sklearn.ensemble.RandomForestClassifier(max_depth=d)
    classifier_score_list.append(sklearn.model_selection.cross_val_score(classifier, X, y, cv=10, n_jobs=4).mean())
plot(d_list, classifier_score_list)
```

```python
#max_features
classifier_score_list = []
f_list = range(1,65)
for f in f_list:
    classifier = sklearn.ensemble.RandomForestClassifier(max_features=f)
    classifier_score_list.append(sklearn.model_selection.cross_val_score(classifier, X, y, cv=10, n_jobs=4).mean())
plot(f_list, classifier_score_list)
```
